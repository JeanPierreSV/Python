### **Variables locales y globales**

```python
# Una variable local no puede modificar a una global
def palindromo(frase):
  frase = frase.replace(" ","") # variable local
  # Reemplazamos el espacio por el vacio
  print(frase)
  return frase == frase[::-1]
  
frase = "anita lava la tina" # Variable global
print(frase)

resultado = palindromo(frase)
print(frase)
print(resultado)
```

```python
animal = "Leon"  #Variable declarada fuera de la funcion
def mostrar():
  global animal  # Estamos creando o revalidando una variable a global
  animal = "Ballena" #Variable local
  print(animal)

print(animal)
mostrar()
print(animal)
```

### **Decoradores**

```python
# Un decorador nos va a permitir otorgarle mayor funcionalidad a una funcion concreta
# Un decorador recibe como parametro una funcion para dar como salida una nueva funcion
def decorador(is_valid = True):
    def _decorador(func):
        def before_action():
            print("Vamos a ejecutar esta funcion")
        def after_action():
            print("Se ejecuto la funcion")
        def nueva_funcion(*args,**kwargs):
            if is_valid:
                before_action()
            resultado = func(*args,**kwargs) #La ejecuto
            after_action()
            return resultado
        return nueva_funcion
    return _decorador

@decorador() #@decorador(is_valid=True)
def suma(num_uno,num_dos):
    return num_uno + num_dos
resultado = suma(80,17)
print(resultado)
```

### **DOCString**

Documentar funciones

```python
# Queremos documentar "generador" para futuros programadores
# Colocamos la documentacion en laprimera linea dentro de la funcion dentro de """... """

def generador(*agrs):
    """ Recibe n cantidades de numeros y regresa \n el numero ademas de regresar True o False """
    for valor in agrs:
        yield valor*10, True if valor>5 else False

# __doc__ (atributo)
# En python ,las funciones son clases y al ser una clase tiene un atriubto doc
nombre = generador.__name__
documentacion = generador.__doc__
print(nombre, " : ")
print(documentacion)
```

## **13.Excepciones**

Errores que se presentan en el script, es un gestor de errores colocamos codigo sensibles ha arrojar un error entre try y  except

```python
try:
  print(2/0) #Se intenta ejecutar
except:
  print("no es posible la division")
  #Se ejecutara cuando try falle
```

```python
try:
  print(2/0) #Se intenta ejecutar
except ZeroDivisionError: # Tipo de error
  print("no es posible la division")
```

```python
try:
  print(2/0) #Se intenta ejecutar
except ZeroDivisionError as ex: 
  print(ex) # imprimir el tipo de error
  print("no es posible la division")
```

```python
try:
  lista = [1,2]
  print(lista[9])
except IndexError as ex:
  print(ex)
  print("No es posible obtener el valor 9")
except ZeroDivisionError as ex:
  print(ex)
  print("no es posible la division")
```
